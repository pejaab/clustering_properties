Empty Table Creator -> create rows with ids

```python
def create_index(row_num, prefix=None, seed=123456789):
    np.random.seed(seed)
    possible_nums = row_num*10
    rows = np.random.choice(possible_nums, row_num)
    rows = np.sort(rows)
    rows = rows.reshape(row_num, 1)
    prefixed_rows = rows.copy()
    if prefix:
        prefixed_rows = np.core.defchararray.add(prefix, rows.astype(str))
    result = np.append(rows, prefixed_rows, axis=1)
    return result
```

(Random Label Assigner -> give rows labels ("ground truth"))

```python
def create_labels(array, label_1, label_2, *args, seed=123456789):
    np.random.seed(seed)
    probs = []
    labels = []
    labels.append(label_1[0])
    probs.append(label_1[1])
    labels.append(label_2[0])
    probs.append(label_2[1])
    if args:
        for i in range(len(args)):
            label, prob = args[i]
            labels.append(label)
            probs.append(prob)
    scaled = np.array(probs)/sum(probs)
    np.random.shuffle(array)
    size = len(array)
    array_parts = np.rint(size*scaled)
    array_parts = array_parts.astype(int)
    print(array_parts, sum(array_parts), size)
    label_list = []
    for i in range(len(array_parts)):
        label_list += [labels[i]]*array_parts[i] 
    label_list = np.array(label_list).reshape(300,1)
    array = np.append(array, label_list, axis=1)
    df = pd.DataFrame(array[:,1:], index=array[:,0])
    df.index = df.index.astype(int)
    df = df.sort_index()
    return df
```

```python
label_array = create_labels(ind, ('cluster 1', 0.5), ('cluster 2', 0.5))
```

```python
def create_dimension_using_label_based_distribution(df, column, label, dim_name, dist_func, *args):
    if not dim_name in df.columns:
        df[dim_name] = 0
    df[dim_name] = df.apply(lambda row: row[dim_name] if row[column] != label else dist_func(*args), axis=1)
    return df
```

```python
def create_dimension_using_mathematical_function(df, dim_name, math_func):
    df[dim_name] = df.apply(math_func, axis=1)
    return df
```
